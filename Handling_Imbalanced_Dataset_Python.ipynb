Imbalanced Dataset:

A dataset with an unequal class distribution

```python
import pandas as pd
import numpy as np
```

```python
credit_df = pd.read_csv('/content/credit_data.csv')
```

```python
credit_df.head()
```

```python
credit_df.tail()
```

```python
# distribution of 2 classes
credit_df['Class'].value_counts()
```

This is Highly Imbalanced Dataset

0 --> Legit Transaction

1 --> Fraudulent Transaction

```python
# separating the legit and fraudulent Transactions
legit = credit_df[credit_df.Class == 0]
fraud = credit_df[credit_df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling

Building a sample dataset containing similar distribution of legit and fraud transaction

Number of fraudulent transactions = 282

```python
legit_sample = legit.sample(n=282)
```

```python
legit_sample.shape
```

Concantenate 2 Dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```
