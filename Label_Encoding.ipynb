Label Encoding:
- Converting the labels into numeric form

```python
# import the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
cancer_df = pd.read_csv('/content/breast_cancer_data.csv')
```

```python
cancer_df.head()
```

```python
cancer_df.shape
```

```python
# finding the count of different labels
cancer_df['diagnosis'].value_counts()
```

```python
# load label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_df.diagnosis)
```

```python
# appending the labels to the dataframe
cancer_df['target']=labels
```

```python
cancer_df.head()
```

0 --> Benign
1 --> Malignant

```python
cancer_df['target'].value_counts()
```

Label Encoding of Iris Data

```python
iris_df = pd.read_csv('/content/iris_data.csv')
```

```python
iris_df.head()
```

```python
iris_df['Species'].value_counts()
```
